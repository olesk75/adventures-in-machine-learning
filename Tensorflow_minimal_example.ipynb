````
import tensorflow as tf
import numpy as np
import keras
````

````
print(tf.__version__)
````

Initial hello world equivalent - single node network

````
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
````

The *loss* function determines how big the error is.
The *optimizer* function picks the next guess

````
model.compile(optimizer='sgd', loss='mean_squared_error')
````

The data, two arrays of numbers, 
```Y = 3x + 1```



````
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype=float)
````

We try to fit the two numpy arrays, 500 iterations

````
model.fit(xs, ys, epochs=500)
````

Finally we have a trained model and can now make a prediction.

Remember: ```Y = 3x + 1```

````
print(model.predict([12.0]))
````
